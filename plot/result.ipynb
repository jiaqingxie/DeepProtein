```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(20, 7), sharey=False)  # Changed sharey to False

models = ['CNN', 'Transformer', 'GCN', 'NeuralFP']
metrics = ['PR-AUC', 'R^2', 'MAE']

pr_auc_values = [[90.88, 78.38, 83.43, 81.14],
                 [0.493, 0.425, 0.394, 0.100],
                 [0.0745, 0.1216, 0.1274, 0.1243]]
errors = [[0.31, 0.25, 0.10, 0.06],
          [0.015, 0.021, 0.006, 0.054],
          [0.0005, 0.0020, 0.0019, 0.0001]]
# Define bar positions
x = np.arange(len(models))

# Titles for each subplot
dataset_name = ['SubCellular', 'PPI Affinity', 'CRISPR Leenay']
colors = ['red', 'blue', 'purple', 'green']

for i, ax in enumerate(axes):
    # Adding a grid and background color (only on y-axis)
    ax.set_facecolor('#eaeaf2')
    ax.grid(True, axis='y', color='white', linestyle='-', linewidth=1.5, alpha=0.5)  # Apply grid only on y-axis

    # Plotting the bars for each subplot after the grid
    ax.bar(x, pr_auc_values[i], yerr=errors[i], color=colors, alpha=1, edgecolor='black', capsize=5, width=0.6, zorder=3)

    # Customizing the plot
    ax.set_xticks(x)
    ax.set_xticklabels(models, fontdict={'family': 'serif', 'size': 16})  # Increased font size
    ax.set_title(dataset_name[i], fontdict={'family': 'serif', 'size': 20})  # Place dataset names as titles
    
    ax.set_ylabel(metrics[i], fontdict={'family': 'serif', 'size': 18})  # Generalized label for y-axis

    # Set y-limits based on the values
    ax.set_ylim(min(pr_auc_values[i]) - 0.05, max(pr_auc_values[i]) + 0.1)

    # Increase the font size of the y-axis tick labels
    ax.tick_params(axis='y', labelsize=16) 

    # Ensure bars are drawn on top of the grid
    for spine in ax.spines.values():
        spine.set_zorder(4)  # Ensure axis lines are above grid if needed
    ax.grid(True, color='white', linestyle='-', linewidth=1.5, alpha=0.5)

plt.tight_layout()
# plt.savefig("example_metrics.eps")  # Saving the plot as EPS
plt.savefig("example_metrics.png")  # Saving the plot as PNG
plt.show()
```
